```python
import os

def extract_relevant_text(file_path):
    with open(file_path, "r") as f:
        contents = f.read()
        start_title = contents.find("<TITLE>")
        end_title = contents.find("</TITLE>")
        start_text = contents.find("<TEXT>")
        end_text = contents.find("</TEXT>")
        title = contents[start_title + 7:end_title]
        text = contents[start_text + 6:end_text]
        final_contents = title + " " + text
        return final_contents

file_contents = {}

directory = path
for filename in os.listdir(directory):
    file_path = os.path.join(directory, filename)
    if os.path.isfile(file_path):
        file_contents[filename] = extract_relevant_text(file_path)

sample_files = list(file_contents.keys())[:5]

for filename in sample_files:
    print("Before:")
    with open(os.path.join(directory, filename), "r") as f:
        print(f.read())
    print("\n\n")
    print("After:")
    print(file_contents[filename])
    print("\n")
```

```python
import os

# Looping through all files in the folder
for filename in os.listdir(path):
    # Open the file and read its contents
    with open(os.path.join(path, filename), 'r') as file:
        contents = file.read()
        
    # Extract the contents between the <TITLE>...</TITLE> and <TEXT>...</TEXT> tags
    start_title = contents.find("<TITLE>") + len("<TITLE>")
    end_title = contents.find("</TITLE>")
    start_text = contents.find("<TEXT>") + len("<TEXT>")
    end_text = contents.find("</TEXT>")
    
    # Concatenate the two strings using a blank space
    relevant_text = contents[start_title:end_title] + ' ' + contents[start_text:end_text]

    # Write the relevant text back to the same file
    with open(os.path.join(path, filename), 'wb') as file:
        file.write(b"")
    with open(os.path.join(path, filename), 'w') as file:
        file.write(relevant_text)
        
# Print contents of 5 sample files before and after performing the operation
for i in range(5):
    filename = os.listdir(path)[i]
    with open(os.path.join(path, filename), 'r') as file:
        contents = file.read()
        print(f'Before: {contents[:100]}...\nAfter: {contents[:100]}...\n')
```

```python
import os
import string
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def preprocess_text(text):
    text = text.lower()
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words("english"))
    tokens = [token for token in tokens if token not in stop_words and token not in string.punctuation]
    tokens = [token for token in tokens if token.strip()]
    return " ".join(tokens)

sample_files = list(file_contents.keys())[:5]

for filename in sample_files:
    print("Before lowercasing:")
    print(file_contents[filename])
    file_contents[filename] = file_contents[filename].lower()
    print("After lowercasing:")
    print(file_contents[filename])
    print("\n")

for filename in sample_files:
    print("Before tokenization:")
    print(file_contents[filename])
    file_contents[filename] = word_tokenize(file_contents[filename])
    print("After tokenization:")
    print(file_contents[filename])
    print("\n")

for filename in sample_files:
    print("Before removing stopwords:")
    print(file_contents[filename])
    stop_words = set(stopwords.words("english"))
    file_contents[filename] = [token for token in file_contents[filename] if token not in stop_words]
    print("After removing stopwords:")
    print(file_contents[filename])
    print("\n")

for filename in sample_files:
    print("Before removing punctuations:")
    print(file_contents[filename])
    file_contents[filename] = [token for token in file_contents[filename] if token not in string.punctuation]
    print("After removing punctuations:")
    print(file_contents[filename])
    print("\n")

for filename in sample_files:
    print("Before removing blank space tokens:")
    print(file_contents[filename])
    file_contents[filename] = [token for token in file_contents[filename] if token.strip()]
    print("After removing blank space tokens:")
    print(file_contents[filename])
    print("\n")
```
